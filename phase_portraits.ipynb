```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Define the Lorenz system
def lorenz(t, X, sigma, rho, beta):
    x, y, z = X
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return [dxdt, dydt, dzdt]

# Parameters
sigma = 10
rho = 28
beta = 8 / 3

# Initial conditions
initial_conditions = [0, 2, 3]  # Modified initial conditions

# Time span
t_span = (0, 50)
t_eval = np.linspace(t_span[0], t_span[1], 10000)

# Solve the system
sol = solve_ivp(lorenz, t_span, initial_conditions, args=(sigma, rho, beta), t_eval=t_eval)

# Extract the solutions
x, y, z = sol.y

# Plot the XY portrait
plt.figure(figsize=(12, 4))
plt.subplot(131)
plt.plot(x, y, color='b', alpha=0.7, linewidth=0.5)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Lorenz System XY Portrait')
plt.grid(True)

# Plot the YZ portrait
plt.subplot(132)
plt.plot(y, z, color='r', alpha=0.7, linewidth=0.5)
plt.xlabel('Y')
plt.ylabel('Z')
plt.title('Lorenz System YZ Portrait')
plt.grid(True)

# Plot the ZX portrait
plt.subplot(133)
plt.plot(z, x, color='g', alpha=0.7, linewidth=0.5)
plt.xlabel('Z')
plt.ylabel('X')
plt.title('Lorenz System ZX Portrait')
plt.grid(True)

plt.tight_layout()
plt.show()

# Plot the 3D phase portrait
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, color='b', alpha=0.7, linewidth=0.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Lorenz System 3D Phase Portrait')
plt.show()
```

```python
# Define the Rssler system
def rossler(t, X, a, b, c):
    x, y, z = X
    dxdt = -y - z
    dydt = x + a * y
    dzdt = b + z * (x - c)
    return [dxdt, dydt, dzdt]

# Parameters
a = 0.2
b = 0.2
c = 5.7

# Initial conditions
initial_conditions = [10.0, 0.0, 10.0]

# Time span
t_span = (0, 100)
t_eval = np.linspace(t_span[0], t_span[1], 10000)

# Solve the system
sol = solve_ivp(rossler, t_span, initial_conditions, args=(a, b, c), t_eval=t_eval)

# Extract the solutions
x, y, z = sol.y

# Plot the XY portrait
plt.figure(figsize=(12, 4))
plt.subplot(131)
plt.plot(x, y, color='b', alpha=0.7, linewidth=0.5)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Rssler System XY Portrait')
plt.grid(True)

# Plot the YZ portrait
plt.subplot(132)
plt.plot(y, z, color='r', alpha=0.7, linewidth=0.5)
plt.xlabel('Y')
plt.ylabel('Z')
plt.title('Rssler System YZ Portrait')
plt.grid(True)

# Plot the ZX portrait
plt.subplot(133)
plt.plot(z, x, color='g', alpha=0.7, linewidth=0.5)
plt.xlabel('Z')
plt.ylabel('X')
plt.title('Rssler System ZX Portrait')
plt.grid(True)

plt.tight_layout()
plt.show()

# Plot the 3D phase portrait
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, color='b', alpha=0.7, linewidth=0.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Rssler System 3D Phase Portrait')
plt.show()
```

```python
# Define the Aizawa system
def aizawa(t, X, a, b, c, d, e, f):
    x, y, z = X
    dxdt = (z - b) * x - d * y
    dydt = d * x + (z - b) * y 
    dzdt = c + a * z - (z ** 3 / 3) - (x ** 2 + y ** 2) * (1 + e * z) + f * z * x ** 3
    return [dxdt, dydt, dzdt]

# Parameters
a = 0.95
b = 0.7
c = 0.6
d = 3.5
e = 0.25
f = 0.1 

# Initial conditions
initial_conditions = [0.1, 1.0, 0.01]

# Time span
t_span = (0, 100)
t_eval = np.linspace(t_span[0], t_span[1], 10000)

# Solve the system
sol = solve_ivp(aizawa, t_span, initial_conditions, args=(a, b, c, d, e, f), t_eval=t_eval)

# Extract the solutions
x, y, z = sol.y

# Plot the XY portrait
plt.figure(figsize=(12, 4))
plt.subplot(131)
plt.plot(x, y, color='b', alpha=0.7, linewidth=0.5)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Aizawa System XY Portrait')
plt.grid(True)

# Plot the YZ portrait
plt.subplot(132)
plt.plot(y, z, color='r', alpha=0.7, linewidth=0.5)
plt.xlabel('Y')
plt.ylabel('Z')
plt.title('Aizawa System YZ Portrait')
plt.grid(True)

# Plot the ZX portrait
plt.subplot(133)
plt.plot(z, x, color='g', alpha=0.7, linewidth=0.5)
plt.xlabel('Z')
plt.ylabel('X')
plt.title('Aizawa System ZX Portrait')
plt.grid(True)

plt.tight_layout()
plt.show()

# Plot the 3D phase portrait
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, color='b', alpha=0.7, linewidth=0.5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Aizawa System 3D Phase Portrait')
plt.show()
```
